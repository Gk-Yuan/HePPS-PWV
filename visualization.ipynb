```python
# Import required libraries
import matplotlib.pyplot as plt
import numpy as np
from CML_dataset import CMLDataset

# Load the CML dataset
ds = CMLDataset(target_fs=180, target_length=150)
print(f"Loaded {len(ds)} segments from the dataset")
print(f"Available labels: {set(ds.labels)}")
```

```python
# Plot 5 continuous segments
fig, axes = plt.subplots(5, 2, figsize=(15, 12))
fig.suptitle('5 Continuous Segments - Wrist and Finger Channels', fontsize=16)

# Take the first 5 segments
for i in range(5):
    if i < len(ds.segments):
        segment = ds.segments[i]
        label = ds.labels[i]
        sbp = ds.sbp_list[i]
        dbp = ds.dbp_list[i]
        
        # Plot wrist channel (left column)
        axes[i, 0].plot(segment[:, 0], color='tab:red', linewidth=1.5)
        axes[i, 0].set_title(f'Segment {i+1} - Wrist\nLabel: {label}, SBP: {sbp:.1f}, DBP: {dbp:.1f}')
        axes[i, 0].set_ylabel('Normalized Amplitude')
        axes[i, 0].grid(True, alpha=0.3)
        
        # Plot finger channel (right column)
        axes[i, 1].plot(segment[:, 1], color='tab:blue', linewidth=1.5)
        axes[i, 1].set_title(f'Segment {i+1} - Finger\nLabel: {label}, SBP: {sbp:.1f}, DBP: {dbp:.1f}')
        axes[i, 1].set_ylabel('Normalized Amplitude')
        axes[i, 1].grid(True, alpha=0.3)
        
        # Add x-label only to bottom plots
        if i == 4:
            axes[i, 0].set_xlabel('Sample Index')
            axes[i, 1].set_xlabel('Sample Index')

plt.tight_layout()
plt.show()
```

```python
# Plot 5 continuous segments overlaid for comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot wrist channels overlaid
for i in range(5):
    if i < len(ds.segments):
        segment = ds.segments[i]
        label = ds.labels[i]
        ax1.plot(segment[:, 0], alpha=0.7, linewidth=2, label=f'Seg {i+1} ({label})')

ax1.set_title('Wrist Channel - 5 Continuous Segments Overlaid')
ax1.set_xlabel('Sample Index')
ax1.set_ylabel('Normalized Amplitude')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot finger channels overlaid
for i in range(5):
    if i < len(ds.segments):
        segment = ds.segments[i]
        label = ds.labels[i]
        ax2.plot(segment[:, 1], alpha=0.7, linewidth=2, label=f'Seg {i+1} ({label})')

ax2.set_title('Finger Channel - 5 Continuous Segments Overlaid')
ax2.set_xlabel('Sample Index')
ax2.set_ylabel('Normalized Amplitude')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Display statistics for the 5 segments
print("\nSegment Statistics:")
print("=" * 60)

for i in range(5):
    if i < len(ds.segments):
        segment = ds.segments[i]
        label = ds.labels[i]
        sbp = ds.sbp_list[i]
        dbp = ds.dbp_list[i]
        
        wrist_stats = {
            'mean': np.mean(segment[:, 0]),
            'std': np.std(segment[:, 0]),
            'min': np.min(segment[:, 0]),
            'max': np.max(segment[:, 0])
        }
        
        finger_stats = {
            'mean': np.mean(segment[:, 1]),
            'std': np.std(segment[:, 1]),
            'min': np.min(segment[:, 1]),
            'max': np.max(segment[:, 1])
        }
        
        print(f"Segment {i+1} - Label: {label}")
        print(f"  Blood Pressure: SBP={sbp:.1f}, DBP={dbp:.1f}")
        print(f"  Wrist  - Mean: {wrist_stats['mean']:.4f}, Std: {wrist_stats['std']:.4f}, Range: [{wrist_stats['min']:.4f}, {wrist_stats['max']:.4f}]")
        print(f"  Finger - Mean: {finger_stats['mean']:.4f}, Std: {finger_stats['std']:.4f}, Range: [{finger_stats['min']:.4f}, {finger_stats['max']:.4f}]")
        print(f"  Shape: {segment.shape}")
        print()

# Show overall dataset statistics
print(f"\nDataset Overview:")
print(f"Total segments: {len(ds.segments)}")
print(f"Unique labels: {len(set(ds.labels))}")
print(f"Label distribution: {dict(zip(*np.unique(ds.labels, return_counts=True)))}")
print(f"SBP range: {np.min(ds.sbp_list):.1f} - {np.max(ds.sbp_list):.1f}")
print(f"DBP range: {np.min(ds.dbp_list):.1f} - {np.max(ds.dbp_list):.1f}")
```
